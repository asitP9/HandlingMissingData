### Removing the column with maximum missing values

```python
for col in cols:
#     missing_pct=np.mean(df[col].isnull())
#     print("\t {} - {}%".format(col, missing_pct*100))

    missing=df[col].isnull()
    print(missing)
    num_missing=np.sum(missing)
    if num_missing>0:
        df['{}_ismissing'.format(col)]=missing
```

```python
ismiss_cols=[col for col in df.columns if "ismissing" in col]
```

```python
ismiss_cols
```

```python
df['num_missing']=df[ismiss_cols].sum(axis=1)
df
```

```python
row_missing=df[df['num_missing']>4].index
```

```python
row_missing
```

```python
df.to_csv("modified_df.csv", index=False)
```
